```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('D:/Data Science Class/Data science 3/my/data/housing-data.txt')
df.head()
```

```python
df.plot.scatter(x = 'size', y = 'price', s = 10)
```

```python
def calculating_cost(x, y, Tet):
    m = len(y)  # Number of training examples
    cost = np.sum(np.square(x.dot(Tet) - y)) / (2.0 * m)

    return cost
```

```python

# Extract the feature variable 'size' from the DataFrame 'df'
X = df['size']

# Extract the target variable 'price' from the DataFrame 'df'
y = df['price']

# Initialize the parameter vector 'theta' as a 1x1 array filled with zeros
Tet = np.zeros((1, 1))

# Reshape the shape of 'X' to be a column vector using the 'reshape()' function
X = X.values.reshape((len(X), 1))

# Reshape the shape of 'y' to be a column vector using the 'reshape()' function
y = y.values.reshape((len(y), 1))

# Call the 'compute_cost()' function to calculate the cost using 'X', 'y', and 'theta'
cost = calculating_cost(X, y, Tet)

# Print the calculated cost
print(cost)
```

```python
def gradient_descent(X, y, Tet, ALP, range_number):
    m = len(y)
    J = []

    for i in range(range_number):
        #Tet = Tet - (ALP / (m * np.dot((np.dot(X,theta)-y),X.T)))
        Tet -= ALP / m * ((X.dot( Tet) - y).T.dot(X))
        cost = calculating_cost(X, y, Tet)
        J.append(cost)

    return J
```

```python
ALP = 0.00007
range_number = 50
costs = gradient_descent(X,y,Tet,ALP, range_number)

plt.plot(costs, color = 'b')
plt.show()
```
